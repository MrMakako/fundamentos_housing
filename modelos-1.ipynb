```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from  sklearn import preprocessing
from sklearn.linear_model import LinearRegression

```

```python
x=[i for i in range(50)]
y=[ 2*num+3 for num in x ]

x=pd.Series(x).to_numpy().reshape(-1,1)
y=pd.Series(y).to_numpy().reshape(-1,1)


model =LinearRegression()

model.fit(x,y)

print(model.score(x,y))

print("coefieciente ",model.coef_)
print("Intercepto en y:",model.intercept_)
print("prediccion f(100) y f(101)",model.predict([[100],[101]]))

plt.scatter(x,y,color="r")
```

```python
#Generar modelo exponencial fx(x)=3x**2+2
x= [i for i in range(-25,25)]
y= [3*el**2 +2 for el in x]
#Transfromacion
x=pd.Series(x).to_numpy().reshape(-1,1)
y=pd.Series(y).to_numpy().reshape(-1,1)
#crear un objeto con caractereisticas polinomiales de grado 2
poly_features=preprocessing.PolynomialFeatures(degree=2)
x_poly=poly_features.fit_transform(x)


#Creae modelo
model=LinearRegression()
model.fit(x_poly,y)


#imrpimri coefciiente
print("Coeficiente ",model.coef_)
# elmultimo elemenot es lo que nos interesa 
#donde 3 es elemento que habiamos definido 3x


#imprimamos el intercepto en y

print("Intercepto en y:", model.intercept_)
#3x**2 +2

#predeci f(100)
#para general el modello expoencial generamos un modelo con paramaetros de grado 2 
print("prediccion f(100)", model.predict([[1,100,100**2]]))

#predic f(-1)
print("prediciton f(-1)",model.predict([[1,-1,1]]))

plt.scatter(x,y,color="purple")
```
